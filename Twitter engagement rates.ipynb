```python
# import libraries

import pandas as pd
import tweepy
import time
import numpy as np
from fractions import Fraction
import os
```

```python
# opening tweet file

df = pd.read_excel(r'ShazaFK.xlsx')
df
```

```python
# using tweepy to make a client

client = tweepy.Client(bearer_token='token')
```

```python
df = df.drop_duplicates(subset=["text"], keep='first')
df = df.reset_index()
len(df['id'])
```

```python
# extracting tweets public_metrics from the client 

gathered_tweets = []

for i in range(len(df['id'])):
    
        for response in tweepy.Paginator(client.get_tweets,ids=df['id'][i],tweet_fields=['public_metrics','lang']):
            
            time.sleep(3)
            gathered_tweets.append(response)
```

```python
# saving public_metrics

result = []

for response in gathered_tweets:
    if response.data != None:
        for tweet in response.data:
            
            result.append({'author_id': tweet.author_id,
                   'tweet_id': tweet.id,
                   'retweets': tweet.public_metrics['retweet_count'],
                   'replies': tweet.public_metrics['reply_count'],
                   'likes': tweet.public_metrics['like_count'],
                   'quotes': tweet.public_metrics['quote_count'],
                     'lang': tweet.lang
                   }) 
    else:
        
        result.append({'author_id': 'None',
                   'tweet_id': 'None',
                   'retweets': 0,
                   'replies': 0,
                   'likes': 0,
                   'quotes': 0,
                     'lang': 'None'
                   })
       
```

```python
result
```

```python
# converting id to string variable --> to save it in excel file (as excel rounds off long int values)

df['id'] = df['id'].map(str)
df
```

```python
retweets = []
replies = []
likes = []
lang = []

for i in range(len(result)):
    
    retweets.append(result[i]['retweets'])
    replies.append(result[i]['replies'])
    likes.append(result[i]['likes'])
    lang.append(result[i]['lang'])
    
replies
```

```python
# making a new df that will be saved as excel file

new_df = df.drop(['author_id', 'in_reply_to_user_id', 'lang','author','entities','referenced_tweets','media','geo'], axis=1)
new_df
```

```python
new_df['retweets'] = retweets
new_df['replies'] = replies
new_df['likes'] = likes
new_df['lang'] = lang

new_df
```

```python
new_df.isnull().sum()
```

```python
new_df.drop(new_df[(new_df['retweets'] ==0) & (new_df['lang'] == 'None')].index, inplace=True)
new_df
```

```python
count_1 = (new_df['retweets'] == 0).sum()
count_2 = (new_df['likes'] == 0).sum()
count_3 = (new_df['replies'] == 0).sum()

print('Count of zeros in Column  retweets : ', count_1)
print('Count of zeros in Column  likes : ', count_2)
print('Count of zeros in Column  replies : ', count_3)
```

```python
Fraction(count_2,len(new_df['id']))
```

```python
matching = [s for s in new_df['text'] if "RT" in s]
len(matching)
```

```python
weights = []

weights.append(( (len(new_df['text'])/3) + (count_2-count_1) )/len(new_df['text']))
weights.append(( (len(new_df['text'])/3) - ((count_2-count_1)/2) )/len(new_df['text']))
weights.append(( (len(new_df['text'])/3) - ((count_2-count_1)/2) )/len(new_df['text']))
weights[0]+weights[1]+weights[2]
```

```python
if weights[0] > 1:
    weights[0] = 1
    weights[1] = 0
    weights[2] = 0
```

```python
weighted_avg = weights[0]*new_df['retweets'] + weights[1]*new_df['replies'] + weights[2]*new_df['likes']
new_df['score'] = weighted_avg
new_df
```

```python
new_df['norm_score'] = new_df['score']/max(new_df['score'])
new_df
```

```python
min(new_df['norm_score'])
```

```python
round((0.1*len(new_df['text'])))
```

```python
top_10_df = new_df.nlargest(n=round((0.1*len(new_df['text']))), columns=['norm_score'], keep='all')
low_10_df = new_df.nsmallest(n=round((0.1*len(new_df['text']))), columns=['norm_score'], keep='all')
low_10_df
```

```python
sort_df = new_df.sort_values('norm_score')
avg_e1 = (len(new_df['text'])/2)-(round((0.1*len(new_df['text'])))/2)
avg_e2 = (len(new_df['text'])/2)+(round((0.1*len(new_df['text'])))/2)
print(avg_e2-avg_e1)
```

```python
avg_10_df = sort_df[int(avg_e1):int(avg_e2)]
avg_10_df
```
