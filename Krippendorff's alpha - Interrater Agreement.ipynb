```python
df_hassan = pd.read_excel('Cohen Kappa .xlsx',0)
df_shayan = pd.read_excel('Cohen Kappa .xlsx',1)
df_ibrahim = pd.read_excel('Cohen Kappa .xlsx',2)
df_maham = pd.read_excel('Cohen Kappa .xlsx',3)
df_shumail = pd.read_excel('Cohen Kappa .xlsx',4)
```

```python
df_hassan['annotator'] = 'hassan'
df_shayan['annotator'] = 'shayan'
df_ibrahim['annotator'] = 'ibrahim'
df_maham['annotator'] = 'maham'
df_shumail['annotator'] = 'shumail'
```

```python
df = pd.concat([df_hassan, df_shayan, df_ibrahim, df_maham, df_shumail], axis=0)
df
```

```python
simpledorff.calculate_krippendorffs_alpha_for_df(df,experiment_col='links',annotator_col='annotator',class_col=['classification category','classification source','open in twitter [yes/no]','verified from twitter only [yes/no]'])
```

```python
c1 = simpledorff.calculate_krippendorffs_alpha_for_df(df,experiment_col='links',annotator_col='annotator',class_col=['classification category'])
c2 = simpledorff.calculate_krippendorffs_alpha_for_df(df,experiment_col='links',annotator_col='annotator',class_col=['classification source'])
c3 = simpledorff.calculate_krippendorffs_alpha_for_df(df,experiment_col='links',annotator_col='annotator',class_col=['open in twitter [yes/no]'])
c4 = simpledorff.calculate_krippendorffs_alpha_for_df(df,experiment_col='links',annotator_col='annotator',class_col=['verified from twitter only [yes/no]'])
```

```python
print(c1)
print(c2)
print(c3)
print(c4)
```

```python
# An α=0.5 means the annotators agreed on 50% of the labels they were expected to disagree on by chance.

# An α=0.8 means the annotators agreed on 80% of the labels they were expected to disagree on by chance.

# An α=−1.0 means the annotators disagreed on 200% of the labels they were expected to disagree on by chance.
```
